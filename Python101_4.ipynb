```python
from collections import Counter

class crimeReport():
    def __init__(self, location, date, crime_category = 'all-crime'):
        self.location = location
        self.date = date
        self.crime_category = crime_category
        self.crimes = self.find_crime()
        
    def find_crime(self):
        crimeURL = 'https://data.police.uk/api/crimes-no-location'
        
        payload = {
           'category' : self.crime_category,
           'force' : self.location,
           'date' : self.date}

        response = requests.get(crimeURL, params= payload) 
        
        if response.status_code == 200:
            return response.json()
        else:
            return None
        
    def reporting_crimse(self):
        crime_list = []
        
        if self.crimes is not None:
            for i in self.crimes:
                crime_list.append(i['category'])
                
            return Counter(crime_list)
        
        
        
```

```python
sr = crimeReport('city-of-london','2024-09')

```

```python
sr.reporting_crimse()
```

```python
sr = crimeReport('nottinghamshire','2024-09')
sr.reporting_crimse()
```

```python
sr = crimeReport('nottinghamshire','2024-09',crime_category ='drugs')
sr.reporting_crimse()
```

```python
class Player():
    def __init__(self,name,pac,sho,pas,dri,deff,phy):
        self.name = name
        self.pac = pac
        self.sho = sho
        self.pas = pas
        self.pac = pac
        self.dri = dri
        self.deff = deff
        self.phy = phy
        
    def skill_ready(self):
        return ','.join([
        str(self.pac),
        str(self.sho),
        str(self.pas),
        str(self.pac),
        str(self.dri),
        str(self.deff),
        str(self.phy),
        str(self.pac)   #If it does not end with the value it started in the code, 
                 ])     #it does not see the final value. so we put the initial variable at the end.
                 
        
    def skill_graph(self):
        graphURL= 'https://image-charts.com/chart'
        payload = {
           'chco' : '3092de',
           'chd' : 't:' + self.skill_ready(),
           'chl' : 'PAC|SHO|PAS|DRI|DEF|PHY',
           'chdl' : self.name,
           'chdlp' : 'b',
           'chs' : '480x480',
           'cht' : 'r',
           'chxt' : 'r',
           'chtt': 'Player Skill',
           'chxl' : '0:|0|20|40|60|80|100',
           'chxt' : 'x',
           'chxr' : '0,0.0,100.0' ,
           'chm' : 'B,AAAAAABB,0,0,0'
        }
        
        response = requests.post( graphURL , data= payload )
        image = Image.open(BytesIO(response.content))
        image.show()
        
    def skill_compare(self, other_player):
        
        graphURL= 'https://image-charts.com/chart'
        payload = {
           'chco' : '3092de,027182',
           'chd' : 't:' + self.skill_ready()+ '|' + other_player.skill_ready(),
           'chl' : 'PAC|SHO|PAS|DRI|DEF|PHY',
           'chdl' : self.name + '|' + other_player.name,
           'chdlp' : 'b',
           'chs' : '480x480',
           'cht' : 'r',
           'chxt' : 'r',
           'chtt': 'Player Skill',
           'chxl' : '0:|0|20|40|60|80|100',
           'chxt' : 'x',
           'chxr' : '0,0.0,100.0' ,
           'chm' : 'B,AAAAAABB,0,0,0|b,0073CFBB,1,0,0'
        }
        response = requests.post( graphURL , data= payload )
        image = Image.open(BytesIO(response.content))
        image.show()
        
```

## Send email with attachment

```python
from email.mime.multipart import MIMEMultipart
from emial.mime.text import MIMEText
from emial.mime.base import MIMEBase
from email import encoders

mail = MIMEMultipart()
mail['From'] = users
mail['To'] = users
mail['Subject'] = 'testing'

mail.attach(MIMEText(message, 'plain'))
attachment_file_name = "testing.jpg"

with(open(attachment_file_name,'rb')) as attachment_file:
    payload = MIMEBase('application','octate-stream')
    payload.set_payload((attachment_file).read())
    encoders.encode_base64(payload)
    
    payload.add_header('Content-Decomposition','attachment',filename = attachment_file_name)
    mail.attach(payload)
    
    mail_str = mail.as_string()
```

```python
port = 465
host = "smtp.gamil.com"

email_server = smtplib.SMTP_SSL(host = host, port = port, context = context)
email_server.login(users, password)
email_server.sendmail(users, recipient, mail_str)
```
